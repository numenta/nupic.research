```python
distr = Normal(0, 1)
import numpy as np

xaxis = torch.arange(0, 1, 0.01, dtype=torch.float32)
Y = []
sd = .05
mu = .5
for x in xaxis:
    y = torch.exp(-(((x-mu)/(2*np.pi*sd))**2))
    Y.append(y)

plt.plot(xaxis.tolist(), Y)
plt.ylim(0, 1)
```

```python
import numpy as np

xaxis = torch.arange(0, 1, 0.001, dtype=torch.float32)
mus = [0, 0.5, 1.0]
sigmas = [0, .1, .5, .8, 1, 2]

fig, axs = plt.subplots(1, 2, dpi=300)

sigma = 2
for mu in mus:
    Y = []
    for x in xaxis:
        y = (1./x) * 1/(sigma*np.sqrt(2*np.pi)) * torch.exp(-1 * (torch.log(x) - mu)**2 / (2*sigma**2))
        Y.append(y)

    axs[0].plot(xaxis.tolist(), Y, label="mu=%.1f" % mu)
    axs[0].legend()
   
mu = 1
for sigma in sigmas:
    Y = []
    for x in xaxis:
        y = (1./x) * 1/(sigma*np.sqrt(2*np.pi)) * torch.exp(-1 * (torch.log(x) - mu)**2 / (2*sigma**2))
        Y.append(y)

    axs[1].plot(xaxis.tolist(), Y, label="sigma=%.1f" % sigma)
    axs[1].legend()    
```
