```python
import torch
from torch.distributions.log_normal import LogNormal
import matplotlib.pyplot as plt
```

```python
distr = LogNormal(0, 1)

xaxis = torch.arange(0, 20, 0.01, dtype=torch.float32)
Y = []
for x in xaxis:
    y = distr.log_prob(7-6*x).exp().item()
    Y.append(y)

plt.plot(xaxis.tolist(), Y)
```
